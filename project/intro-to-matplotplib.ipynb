# Matplotlib workflow

```python
# 1. Import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 2. Prepare data

x = [1,2,3,4]
y = [11,22,33,44]

# 3. Setup plot
fig, ax = plt.subplots(figsize=(5,5))

# 4. Plot data
ax.plot(x,y)

# 5. Customize plot
ax.set(title='Sample plot', xlabel = 'x-axis', ylabel ='y-axis')

# 6. Save and show
fig.savefig('images/sample-plot.png')
```

## Which method should I use? Pyplot or Matplot Object Oriented?

- When plotting something quickly, pyplot method is ok.
- When plotting something more advanced, use Matplot OO method.


```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(x='age',y='chol', c='target',kind='scatter', figsize=(10,6)) # c stands for color
```

```python
## Matplot OO method mixed with Pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(x='age', y='chol', c='target', kind='scatter', ax=ax);

#ax.set_xlim([45,100]) # Change the limitation on the x axis to be from 45 to 100 instead of the default one
```

```python
over_50.target.values
```

# Matplot OO method

```python
# Set the style
plt.style.use('seaborn-v0_8-whitegrid')

# Matplot OO method
fig,ax = plt.subplots(figsize=(10,6))

# Plot the data (scatter is now the name of the figure)
scatter = ax.scatter(x=over_50['age'], 
                     y=over_50['chol'], 
                     c=over_50['target'],
                     cmap='winter'); # This changes the color-scheme

# Customize the plot
ax.set(title='Heart disease and Cholesterol Level',
       xlabel='Age',
       ylabel='Cholesterol');

# Add a legend
ax.legend(*scatter.legend_elements(), title = 'Target');

# Add a horizontal line (mean of the cholesterol)
ax.axhline(over_50['chol'].mean(), 
           linestyle='--');
```

```python
fig, ( (ax0, ax1) ) = plt.subplots(nrows=2,
                                   ncols=1,
                                   figsize=(10,10),
                                   sharex=True)
#Plot the data on ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter')  # This changes the color-scheme

# Customize ax0 plot
ax0.set(title='Heart disease and Cholesterol levels',
        ylabel='Cholesterol')

# Change ax0 x asis limits
ax0.set_xlim([50,80])

# Add a legend on ax0 plot
ax0.legend(*scatter.legend_elements(), title = 'Target');

# Add to  ax0 plot an horizontal meanline of cholesterol
ax0.axhline(over_50['chol'].mean(), 
           linestyle='--');


# Plot the data on ax1
scatter2 = ax1.scatter(x=over_50['age'],
                       y=over_50['thalach'],
                       c=over_50['target'],
                       cmap='winter')
# Customize ax1 plot
ax1.set(title='Heart disease and Max Heart Rate',
        xlabel='Age',
        ylabel='Max Heart Rate')

# Change ax1 x asis limits
ax1.set_xlim([50,80])
# Change ax1 y asis limits
ax1.set_ylim([60,200])

# Add a legend on ax1 plot
ax1.legend(*scatter.legend_elements(), title = 'Target');

# Add to ax1 plot an horizontal line (mean of max heart rate)
ax1.axhline(over_50['thalach'].mean(),
            linestyle='--');

# Add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
